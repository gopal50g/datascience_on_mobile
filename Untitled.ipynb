```python
def plot_subplotted_scatters():
    """ 
    plots of 2 scatterplots as subplots
    """
    return
fig,ax = plt.subplots(dpi=144)
ax.scatter(df['Battery Drain (mAh/day)'],df['Screen On Time (hours/day)'])
ax.set_title('Scatter plot')
ax.set_xlabel('Battery Drain (mAh/day)')
ax.set_ylabel('Screen On Time (hours/day)')
plt.show()
    
```

```python
def plot_subplotted_histograms():
    """plots of a histograms as subplots
    """
    return
fig,ax = plt.subplots()
ax.hist(df['Battery Drain (mAh/day)'], bins=10, alpha=0.7)
ax.set_title('Histogram of Battery Drain')
ax.set_xlabel('Battery Drain (mAh/day)')
ax.set_ylabel('Number')
plt.savefig('hist.jpg', format='jpg')
plt.show()


```
