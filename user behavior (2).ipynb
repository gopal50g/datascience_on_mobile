```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\Anish Kushwaha\\Desktop\\user_behavior_dataset Gopal.csv')
stats=df['Battery Drain (mAh/day)'].describe()
#print('Stats:', (satas,round(2)))

stats = df[['Battery Drain (mAh/day)','Screen On Time (hours/day)']].describe()
print('stats:', (stats.round(2)))

#medians:
median = df[['Battery Drain (mAh/day)','Screen On Time (hours/day)']].median()
print('median:', (median.round(2)))

#skewness:
skewness = df[['Battery Drain (mAh/day)','Screen On Time (hours/day)']].skew()
print('skewness:', (skewness.round(2)))

#kurtosis:
kurtosis = df[['Battery Drain (mAh/day)','Screen On Time (hours/day)']].kurt()
print('kurtosis:', (kurtosis.round(2)))

#correlations:
correlation = df['Battery Drain (mAh/day)'].corr(df['Screen On Time (hours/day)'])
print('correlation between Battery Drain and screen on Time is',correlation.round(2))






```

```python
def plot_subplotted_scatters():
    """ 
    plots of a scatterplots as subplots
    """
    return
fig,ax = plt.subplots(figsize=(8,8), dpi=144)
ax.scatter(df['Battery Drain (mAh/day)'],df['Screen On Time (hours/day)'])

ax.set_title(' battery Drain vs Screen on time')
ax.set_xlabel('Battery Drain (mAh/day)')
ax.set_ylabel('Screen On Time (hours/day)')


plt.savefig('scatter.jpg', format='jpg')
plt.show()
    
```
